```python
from dhash import DHash
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image

logo1 = Image.open('docs/1.png')
hash1 = DHash.calculate_hash(logo1)

logo2 = Image.open('docs/2.png')
hash2 = DHash.calculate_hash(logo2)

logo3 = Image.open('docs/3.png')
hash3 = DHash.calculate_hash(logo3)
```

```python
plt.figure(figsize = (20, 10))

plt.subplot(1, 3, 1)
plt.title(hash1)
plt.imshow(logo1)

plt.subplot(1, 3, 2)
plt.title(hash2)
plt.imshow(logo2)

plt.subplot(1, 3, 3)
plt.title(hash3)
plt.imshow(logo3)

plt.savefig('docs/logo.png')
```
